```python
#EX 4 
a = np.random.randint(10, size=(10, 5))
b = np.random.randint(2, size=10)
dummy0 = np.zeros(shape=(10,5))
dummy1 = np.zeros(shape=(10,5))
row_arr = 0
row_lab = 0

for i in range(0,10,1):
    if b[row_lab] !=0 :
        np.place(dummy0[row_lab],a[row_arr]>0 , np.round(a[row_arr]))
        row_arr += 1
    else:
        np.place(dummy1[row_lab],a[row_arr]>0 , np.round(a[row_arr]))
        row_arr += 1
    row_lab += 1

dummy0 = dummy0[~np.all(dummy0 == 0, axis=1)]
dummy1 = dummy1[~np.all(dummy1 == 0, axis=1)]
print('Default Value = \n',a)
print('label Value =', b)
print('Label That value is 0 \n',dummy0 , '\n','Label That value is 1 \n',dummy1)






```
